```python
import cv2 as cv
import matplotlib.pyplot as plt
#for reading RGB image
import matplotlib.image as mpimg
```

```python
image_path = r"E:\Photoshops\Navami\Navami studies\Github\webbgalaxy.webp"
```

```python
#for reading BGR image
imageArray = cv.imread(image_path)
```

```python
plt.imshow(imageArray)
plt.show()
```

```python
#converting to RGB from BGR
convertedArray = cv.cvtColor(imageArray,cv.COLOR_BGR2RGB)
plt.subplots(figsize=(9,6))
plt.imshow(convertedArray)
plt.show()
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize = (14,10))
ax1.imshow(convertedArray[:,:,0],cmap = "Reds_r")
ax1.set_title("R",size = 20)
ax2.imshow(convertedArray[:,:,1],cmap = "Greens_r")
ax2.set_title("G",size = 20)
ax3.imshow(convertedArray[:,:,2],cmap = "Blues_r")
ax3.set_title("B",size = 20)
ax4.axis("off")
plt.tight_layout()
plt.show()
```
